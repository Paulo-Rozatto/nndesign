```python
import numpy as np
import matplotlib.pyplot as plt
import components.activations as act
```

```python
def plot_activation(fn):
    x = np.array([x / 10.0 for x in range(-100,100)]) # from -10 to 10 with 200 steps
    y = np.array([fn(n) for n in x])
    plt.plot(x, y)
```

```python
plt.figure(figsize=(15, 18))

plotIdx = 1
for name, val in act.__dict__.items():
    if callable(val) and name != "exp":
        ax = plt.subplot(5, 2, plotIdx)
        plt.title(name)
        plot_activation(val)
        plotIdx += 1
plt.show()
```
